```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
def niveis_cinza(image):
  img = cv.cvtColor(image, cv.COLOR_BGR2GRAY) # Image with dimensions reduced to Grayscale
  img = cv.cvtColor(img, cv.COLOR_GRAY2BGR) # Converts black and white colors to RGB (same dimensions). Necessary just for the images display.
  return img


def preto_branco(image):
  img = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
  suave = cv.GaussianBlur(img, (7,7), 0)
  (T, bin) = cv.threshold(suave, 160, 255, cv.THRESH_BINARY) # Image with dimensions reduced to Black and White
  (T, binI) = cv.threshold(suave, 160, 255, cv.THRESH_BINARY_INV) # Image with dimensions reduced to Black and White
  bin = cv.cvtColor(bin, cv.COLOR_GRAY2BGR)  # Converts black and white colors to RGB (same dimensions). Necessary just for the images display.
  binI = cv.cvtColor(binI, cv.COLOR_GRAY2BGR)  # Converts black and white colors to RGB (same dimensions). Necessary just for the images display.
  return bin, binI
```

```python
img = cv.imread('/content/lena.png')

colorida = img
cinza = niveis_cinza(img)
preto_branca1, preto_branca2 = preto_branco(img)

resultado = np.hstack([colorida, cinza, preto_branca1, preto_branca2])
# If the images don't have the same quantity of dimensions, the hstack returns the ValueError below:
# ValueError: all the input arrays must have same number of dimensions, but the array at index 0 has 3 dimension(s) and the array at index 2 has 2 dimension(s)

cv2_imshow(resultado)
```
